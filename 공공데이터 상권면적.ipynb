```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 독립 변수와 종속 변수 설정
X = 서울[['소계', '인구수']]
y = 서울['상업지역']

# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 적용
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# PCA 결과를 데이터프레임으로 변환
X_pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])

# 회귀 분석을 위한 데이터 프레임 생성
pca_data = pd.concat([X_pca_df, y], axis=1)

# 독립 변수와 종속 변수 설정
X_pca = pca_data[['PC1', 'PC2']]
y = pca_data['상업지역']

# 상수 항 추가 (절편을 포함하기 위해)
X_pca = sm.add_constant(X_pca)

# 회귀 모델 적합
model = sm.OLS(y, X_pca).fit()

# 회귀 분석 결과 출력
print(model.summary())

# 회귀 평면 시각화 (2D)
plt.figure(figsize=(12, 8))

# PCA 결과 산점도
plt.scatter(X_pca_df['PC1'], X_pca_df['PC2'], c=y, cmap='viridis', label='Data Points')

# 회귀 직선 시각화 (PCA 결과를 2차원으로 투영)
# PC1과 PC2에 대한 회귀 직선은 추후 직접적 시각화가 어려울 수 있습니다.
# 따라서 회귀 평면의 예측값을 사용하여 투영된 시각화가 가능합니다.
pc1_range = np.linspace(X_pca_df['PC1'].min(), X_pca_df['PC1'].max(), 100)
pc2_range = np.linspace(X_pca_df['PC2'].min(), X_pca_df['PC2'].max(), 100)

PC1_grid, PC2_grid = np.meshgrid(pc1_range, pc2_range)
X_grid = pd.DataFrame({
    'const': 1,
    'PC1': PC1_grid.ravel(),
    'PC2': PC2_grid.ravel()
})

# 예측값 계산 및 numpy 배열로 변환
y_pred_grid = model.predict(X_grid).values.reshape(PC1_grid.shape)

# 회귀 평면의 등고선 시각화
plt.contourf(PC1_grid, PC2_grid, y_pred_grid, cmap='coolwarm', alpha=0.3)

# 자치구 이름 레이블 추가
for i in range(len(서울)):
    plt.text(X_pca_df['PC1'].iloc[i], X_pca_df['PC2'].iloc[i], 서울['자치구별(2)'].iloc[i],
             fontsize=9, ha='right')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA와 회귀 직선 시각화')
plt.colorbar(label='상업지역')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 숫자형 데이터만 선택
서울_numeric = 서울.select_dtypes(include=[np.number])

# 동작구 데이터
dongjak_data_values = 서울_numeric[서울['자치구별(2)'] == '동작구'].values.flatten()

# 유클리드 거리 계산 함수
def euclidean_distance(row1, row2):
    return np.sqrt(np.sum((row1 - row2) ** 2))

# 거리 계산
서울['거리'] = 서울_numeric.apply(lambda row: euclidean_distance(row, dongjak_data_values), axis=1)

# 동작구는 제외하고 정렬
비슷한_자치구 = 서울[서울['자치구별(2)'] != '동작구'].sort_values(by='거리')

# Pandas 출력 옵션 설정
pd.set_option('display.float_format', lambda x: f'{x:.2f}')

print(비슷한_자치구[['자치구별(2)', '거리']])
```
